```python
@jit(nopython = True, nogil = True)
def tak(b_t, initial_imbalance, alpha):

    weighted_count = 0 # Denominator for normalization of EWMAs
    weighted_sum_T = 0 # Nominator for EWMA of duration of a bar
    weighted_sum_imbalance = 0 # Nominator for EWMA of the imbalance of bar

    out = np.zeros(b_t.shape)
    dummy = 0
    imbalance = initial_imbalance
    T = 0
    debug = np.zeros(b_t.shape)

    for i in range(len(b_t)):
        dummy += b_t[i]
        debug[i] = dummy
        T += 1
        if (abs(dummy)>=imbalance):
            out[i] = 1
            weighted_sum_T = T + (1-alpha)*weighted_sum_T
            weighted_sum_imbalance = dummy/(1.0*T) + (1 - alpha) * weighted_sum_imbalance
            weighted_count = 1 + (1-alpha) * weighted_count
            ewma_T = weighted_sum_T/weighted_count
            ewma_imbalance = weighted_sum_imbalance/weighted_count
            imbalance = ewma_T * abs(ewma_imbalance)
            dummy = 0
            T = 0

    return out, debug
```

```python
@jit(nopython=True, nogil = True)   
def imbalance_volume_bars2(volume, initial_imbalance, alpha):

    index = np.zeros(len(volume))
    index_plus = []
    index_minus = []
    temp_plus = 0
    temp_minus = 0
    imbalance = initial_imbalance
    ewma_plus = initial_imbalance
    ewma_minus = -initial_imbalance
    debug = []

    for i in range(len(volume)):
        
        temp_plus += volume[i] * (volume[i] > 0)
        temp_minus += volume[i] * (volume[i] < 0)
    
        if np.abs(temp_plus + temp_minus) > np.abs(imbalance):
            index[i] = np.sign(temp_plus + temp_minus)
            index_plus.append(temp_plus)
            index_minus.append(temp_minus)
            debug.append(imbalance)

            ewma_plus = alpha * temp_plus + (1 - alpha) * ewma_plus
            ewma_minus = alpha * temp_minus + (1 - alpha) * ewma_minus

            #imbalance = ewma_plus - ewma_minus
            #sum1 = 0
            #sum2 = 0
            #for i in range(7):
                #sum1 += index_plus[-7:][i]
                #sum2 += index_minus[-7:][i]
            
            imbalance = ewma_plus + ewma_minus

            temp_plus = 0
            temp_minus = 0
            
    return index, index_plus, index_minus, debug
```

```python
@jit(nopython = True, nogil = True)
def tick_run_bars(b_t, maxi, count_param):
    n = len(b_t)
    count_plus = 0
    count_minus = 0
    signal = 0
    out = np.zeros(n)
    summs = np.zeros(n)

    for i in np.arange(n):
        count_plus += b_t[i] * (b_t[i] > 0)
        count_minus += b_t[i] * (b_t[i] < 0)
        #local_maximum = max([count_plus, - count_minus])
        #summs[i] = local_maximum
        if (max([count_plus, - count_minus]) > maxi):   
        #& ((count_plus/((-count_minus) + 1) > 2) | ((-count_minus)/(count_plus + 1) > 2)):
            maxi = max([count_plus, - count_minus])
            summs[i] = maxi
            signal = i
            counter = 0
        counter += 1
        if (counter > count_param) & (signal != 0):
            temp = b_t[signal : signal + counter]
            out[signal] = 1 if maxi == count_plus else -1
            for i in range(len(temp)):
                count_plus += temp[i] * (temp[i] > 0)
                count_minus += temp[i] * (temp[i] < 0)
            maxi = max([count_plus, - count_minus])
            signal = 0

    return out, summs
```

```python
@jit(nopython = True, nogil = True)
def tick_run_bars2(b_t, initial_maximum, param):
    n = len(b_t)
    sum_t = 0
    temp = []
    out = np.zeros(n)
    maximum = initial_maximum

    for i in np.arange(n):
        sum_t += b_t[i]
        temp.append(sum_t)
        if np.abs(sum_t) > maximum:
            maximum = np.abs(sum_t)
            counter = i
        
        if i - counter > param:
            out[i - counter] = np.sign(sum_t)
            sum_t = 0
            temp2 = temp[counter : i]
            for j in range(i-counter):
                sum_t += temp2[j]
            maximum = initial_maximum
    
    return out, temp
```
